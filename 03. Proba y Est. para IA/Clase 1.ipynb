## Ejercicio 1 

 a.  Cuál es la probabilidad de obtener exactamente 3 cecas en 10 tiradas de una moneda balanceada?

$ P\lbrace{X=i}\rbrace=\begin{pmatrix}N\\i\end{pmatrix}p^{i}(1-p)^{N-i}$ con $i$ éxitos y $N$ tiradas

```python
import math

def binom(n,k):
    return math.factorial(n)/ (math.factorial(k) * math.factorial(n - k))
p = 0.5
i = 3
N = 10 

P = binom(N, i) * p**i * (1-p)**(N-i)

print('P ='+str(P))

```

b. Cuál es la probabilidad de obtener al menos 3 cecas en 10 tiradas de
una moneda cargada donde la probabilidad de ceca es 0.4?

```python
import math

def binom(n,k):
    return math.factorial(n)/ (math.factorial(k) * math.factorial(n - k))
P = 0
p = 0.4
N = 10 

for i in range(4):
    P += binom(N, i) * p**i * (1-p)**(N-i)
print('P ='+str(P))
```

## Ejercicio 4

a. Sean X,Y dos v.a. i.i.d. N(0; 1). Encontrar la expresión de la función
de densidad de probabilidad conjunta.

Siendo $X$, $Y$ dos v.a. i.i.d. normales $\sim\mathcal{N}(0,1)$, se demuestra entonces que para cualquiera sean los  los intervalos $(a,b]$, $(c,d]$ se verifica que $X$ e $Y$ son independientes y entonces:

$P(a<X\leq b,c<Y\leq d)=P(a<X\leq b) P(c<Y\leq d)$

Siendo:

$f_X(x)=\frac{1}{\sqrt{2\pi}}e^{-\frac{x^{2}}{2}} \\
f_Y(y)=\frac{1}{\sqrt{2\pi}}e^{-\frac{y^{2}}{2}}$

Entonces:

$f_{X,Y}(x,y)= \frac{1}{2\pi}e^{-\frac{x^{2}+y^{2}}{2}} $

b. Cuál es la probabilidad de que X > 0.7 y Y < 0.4 simultáneamente?

```python
from scipy.stats import norm

F_x = norm.cdf(0.4)
F_y = 1 - norm.cdf(0.7)

cdf = F_x*F_y
print('F_x,y=', cdf)
```

c. Simular en Octave escribiendo la pdf conjunta o usando funciones de
Octave mvnrnd,mvnpdf,mvncdf

```python
n = 10000

x = np.random.normal(size=n)
y = np.random.normal(size=n)

fig = plt.figure()
ax1 = plt.axes(projection='3d')
plt.title('pdf')
ax1.plot3D(x, y, x*y, 'green', linestyle='none', marker='.')
plt.show()

binorm = np.vstack([x, y])
succesfull_cases = np.sum((binorm.T[:,0]>0.7) & (binorm.T[:,1]<0.4))
print('Probabilidad simulada=',succesfull_cases/len(x))
```
